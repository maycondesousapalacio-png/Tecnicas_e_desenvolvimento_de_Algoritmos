```python
# Sistema de Gerenciamento de Biblioteca

# Lista para armazenar os livros
biblioteca = []

# Dicion√°rio para categorias de livros
categorias = {
    "1": "Fic√ß√£o",
    "2": "N√£o-Fic√ß√£o",
    "3": "T√©cnico",
    "4": "Romance",
    "5": "Fantasia"
}

def menu_principal():
    """Exibe o menu principal do sistema"""
    print("\n" + "="*50)
    print("       SISTEMA DE GERENCIAMENTO DE BIBLIOTECA")
    print("="*50)
    print("1. Adicionar livro")
    print("2. Listar todos os livros")
    print("3. Buscar livro por t√≠tulo")
    print("4. Listar livros por categoria")
    print("5. Estat√≠sticas da biblioteca")
    print("6. Sair")
    print("="*50)

def adicionar_livro():
    """Adiciona um novo livro √† biblioteca"""
    print("\n--- ADICIONAR NOVO LIVRO ---")

    # Estruturas condicionais para valida√ß√£o
    titulo = input("T√≠tulo do livro: ")
    if not titulo:
        print("Erro: T√≠tulo n√£o pode estar vazio!")
        return

    autor = input("Autor: ")
    if not autor:
        print("Erro: Autor n√£o pode estar vazio!")
        return

    # Exibir categorias dispon√≠veis
    print("\nCategorias dispon√≠veis:")
    for codigo, nome in categorias.items():
        print(f"{codigo}. {nome}")

    # Estrutura de repeti√ß√£o para valida√ß√£o da categoria
    while True:
        categoria_codigo = input("\nDigite o n√∫mero da categoria: ")
        if categoria_codigo in categorias:
            categoria = categorias[categoria_codigo]
            break
        else:
            print("Categoria inv√°lida! Tente novamente.")

    # Valida√ß√£o do ano com estrutura condicional
    while True:
        try:
            ano = int(input("Ano de publica√ß√£o: "))
            if ano > 2024 or ano < 1000:
                print("Ano inv√°lido! Digite um ano entre 1000 e 2024.")
            else:
                break
        except ValueError:
            print("Erro: Digite um n√∫mero v√°lido para o ano!")

    # Criar dicion√°rio do livro
    livro = {
        "titulo": titulo,
        "autor": autor,
        "categoria": categoria,
        "ano": ano,
        "disponivel": True
    }

    # Adicionar √† lista
    biblioteca.append(livro)
    print(f"\n‚úÖ Livro '{titulo}' adicionado com sucesso!")

def listar_livros():
    """Lista todos os livros da biblioteca"""
    print("\n--- TODOS OS LIVROS ---")

    # Estrutura condicional para verificar se h√° livros
    if not biblioteca:
        print("Nenhum livro cadastrado na biblioteca.")
        return

    # Estrutura de repeti√ß√£o para percorrer a lista
    for i, livro in enumerate(biblioteca, 1):
        status = "‚úÖ Dispon√≠vel" if livro["disponivel"] else "‚ùå Emprestado"
        print(f"{i}. '{livro['titulo']}' - {livro['autor']} ({livro['ano']})")
        print(f"   Categoria: {livro['categoria']} | Status: {status}")
        print()

def buscar_livro():
    """Busca livros por t√≠tulo"""
    print("\n--- BUSCAR LIVRO ---")
    termo = input("Digite o t√≠tulo ou parte do t√≠tulo: ").lower()

    livros_encontrados = []

    # Estrutura de repeti√ß√£o com condicional para busca
    for livro in biblioteca:
        if termo in livro["titulo"].lower():
            livros_encontrados.append(livro)

    # Estrutura condicional para verificar resultados
    if livros_encontrados:
        print(f"\nüìö {len(livros_encontrados)} livro(s) encontrado(s):")
        for i, livro in enumerate(livros_encontrados, 1):
            status = "‚úÖ Dispon√≠vel" if livro["disponivel"] else "‚ùå Emprestado"
            print(f"{i}. '{livro['titulo']}' - {livro['autor']} | {status}")
    else:
        print("Nenhum livro encontrado com esse t√≠tulo.")

def listar_por_categoria():
    """Lista livros por categoria"""
    print("\n--- LIVROS POR CATEGORIA ---")

    # Exibir categorias
    for codigo, nome in categorias.items():
        print(f"{codigo}. {nome}")

    categoria_codigo = input("\nDigite o n√∫mero da categoria: ")

    # Estrutura condicional para validar categoria
    if categoria_codigo not in categorias:
        print("Categoria inv√°lida!")
        return

    categoria_selecionada = categorias[categoria_codigo]
    livros_categoria = []

    # Buscar livros da categoria selecionada
    for livro in biblioteca:
        if livro["categoria"] == categoria_selecionada:
            livros_categoria.append(livro)

    # Estrutura condicional para verificar resultados
    if livros_categoria:
        print(f"\nüìö Livros na categoria '{categoria_selecionada}':")
        for i, livro in enumerate(livros_categoria, 1):
            status = "‚úÖ Dispon√≠vel" if livro["disponivel"] else "‚ùå Emprestado"
            print(f"{i}. '{livro['titulo']}' - {livro['autor']} ({livro['ano']}) | {status}")
    else:
        print(f"Nenhum livro encontrado na categoria '{categoria_selecionada}'.")

def estatisticas():
    """Exibe estat√≠sticas da biblioteca"""
    print("\n--- ESTAT√çSTICAS DA BIBLIOTECA ---")

    # Estrutura condicional para biblioteca vazia
    if not biblioteca:
        print("Biblioteca vazia!")
        return

    total_livros = len(biblioteca)
    livros_disponiveis = sum(1 for livro in biblioteca if livro["disponivel"])
    livros_emprestados = total_livros - livros_disponiveis

    # Contar livros por categoria usando dicion√°rio
    livros_por_categoria = {}
    for livro in biblioteca:
        categoria = livro["categoria"]
        if categoria in livros_por_categoria:
            livros_por_categoria[categoria] += 1
        else:
            livros_por_categoria[categoria] = 1

    print(f"üìä Total de livros: {total_livros}")
    print(f"‚úÖ Livros dispon√≠veis: {livros_disponiveis}")
    print(f"‚ùå Livros emprestados: {livros_emprestados}")

    print("\nüìö Livros por categoria:")
    for categoria, quantidade in livros_por_categoria.items():
        porcentagem = (quantidade / total_livros) * 100
        print(f"   {categoria}: {quantidade} livro(s) ({porcentagem:.1f}%)")

# Programa principal
def main():
    """Fun√ß√£o principal do programa"""
    print("Bem-vindo ao Sistema de Gerenciamento de Biblioteca!")

    # Estrutura de repeti√ß√£o principal
    while True:
        menu_principal()

        opcao = input("\nDigite a op√ß√£o desejada: ")

        # Estruturas condicionais para menu
        if opcao == "1":
            adicionar_livro()
        elif opcao == "2":
            listar_livros()
        elif opcao == "3":
            buscar_livro()
        elif opcao == "4":
            listar_por_categoria()
        elif opcao == "5":
            estatisticas()
        elif opcao == "6":
            print("\nObrigado por usar nosso sistema! At√© logo! üëã")
            break
        else:
            print("Op√ß√£o inv√°lida! Tente novamente.")

        input("\nPressione Enter para continuar...")

# Executar o programa
if __name__ == "__main__":
    main()
```
