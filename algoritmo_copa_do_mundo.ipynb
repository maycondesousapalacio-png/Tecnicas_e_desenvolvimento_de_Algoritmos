```python
#Exercício 24/10
#Maycon de Sousa Palácio
#Técnicas de Desenvolvimento de Algoritmos

import random
from datetime import datetime

def chaveamento_copa_2026():
    """
    Algoritmo de chaveamento da Copa do Mundo 2026
    Seguindo as regras oficiais da FIFA
    """

    # 1. LISTA DE PAÍSES CLASSIFICADOS POR CONFEDERAÇÃO
    # Baseado no formato expandido para 48 seleções
    paises_por_confederacao = {
        'AFC (Ásia)': ['Japão', 'Irã', 'Coreia do Sul', 'Arábia Saudita', 'Austrália',
                      'Catar', 'Iraque', 'Emirados Árabes', 'Omã', 'China'],

        'CAF (África)': ['Senegal', 'Marrocos', 'Tunísia', 'Argélia', 'Egito',
                        'Nigéria', 'Camarões', 'Gana', 'Costa do Marfim', 'Mali'],

        'CONCACAF (América do Norte)': ['Estados Unidos', 'México', 'Canadá', 'Costa Rica',
                                       'Panamá', 'Jamaica', 'Honduras', 'El Salvador'],

        'CONMEBOL (América do Sul)': ['Brasil', 'Argentina', 'Uruguai', 'Colômbia',
                                     'Equador', 'Peru', 'Chile', 'Paraguai'],

        'OFC (Oceania)': ['Nova Zelândia', 'Ilhas Salomão', 'Papua-Nova Guiné'],

        'UEFA (Europa)': ['França', 'Bélgica', 'Inglaterra', 'Portugal', 'Espanha',
                         'Itália', 'Holanda', 'Alemanha', 'Dinamarca', 'Suíça',
                         'Croácia', 'Sérvia', 'Polônia', 'Ucrânia', 'Suécia']
    }

    # 2. TUPLA COM OS POTES BASEADO NO RANKING FIFA
    # Ordenado por ranking (simulação)
    pote_1 = ('Argentina', 'França', 'Brasil', 'Bélgica', 'Inglaterra', 'Portugal', 'Holanda', 'Espanha')
    pote_2 = ('Itália', 'Croácia', 'Alemanha', 'Dinamarca', 'Suíça', 'Uruguai', 'Colômbia', 'México')
    pote_3 = ('Estados Unidos', 'Senegal', 'Marrocos', 'Japão', 'Irã', 'Coreia do Sul', 'Sérvia', 'Polônia')
    pote_4 = ('Egito', 'Nigéria', 'Austrália', 'Canadá', 'Costa Rica', 'Arábia Saudita', 'Ucrânia', 'Suécia')
    pote_5 = ('Chile', 'Peru', 'Paraguai', 'Tunísia', 'Argélia', 'Camarões', 'Gana', 'Panamá')
    pote_6 = ('Jamaica', 'Honduras', 'Nova Zelândia', 'Catar', 'Iraque', 'Emirados Árabes', 'Omã', 'China')

    # 3. DICIONÁRIO COM RESTRIÇÕES DE CHAVEAMENTO
    restricoes_chaveamento = {
        'CONMEBOL': {'max_por_grupo': 2},  # Máximo 2 países da CONMEBOL por grupo
        'UEFA': {'max_por_grupo': 2},      # Máximo 2 países da UEFA por grupo
        'AFC': {'max_por_grupo': 2},       # Máximo 2 países da AFC por grupo
        'CAF': {'max_por_grupo': 2},       # Máximo 2 países da CAF por grupo
        'CONCACAF': {'max_por_grupo': 2},  # Máximo 2 países da CONCACAF por grupo
        'OFC': {'max_por_grupo': 1}        # Máximo 1 país da OFC por grupo
    }

    # 4. FUNÇÃO PARA IDENTIFICAR A CONFEDERAÇÃO DE UM PAÍS
    def encontrar_confederacao(pais):
        """Retorna a confederação do país"""
        for confederacao, paises in paises_por_confederacao.items():
            if pais in paises:
                return confederacao
        return None

    # 5. ALGORITMO DE CHAVEAMENTO
    print("=== CHAVEAMENTO COPA DO MUNDO 2026 ===\n")

    # Criar grupos de A a H (8 grupos)
    grupos = {}
    for letra in 'ABCDEFGH':
        grupos[f'Grupo {letra}'] = []

    # Lista de potes em ordem de sorteio
    potes = [pote_1, pote_2, pote_3, pote_4, pote_5, pote_6]

    # Contadores para restrições por grupo
    contadores_grupos = {}
    for grupo in grupos.keys():
        contadores_grupos[grupo] = {
            'AFC (Ásia)': 0,
            'CAF (África)': 0,
            'CONCACAF (América do Norte)': 0,
            'CONMEBOL (América do Sul)': 0,
            'OFC (Oceania)': 0,
            'UEFA (Europa)': 0
        }

    # Sorteio dos times
    random.seed(datetime.now().timestamp())  # Semente aleatória

    print("INICIANDO SORTEIO...\n")

    for i, pote in enumerate(potes, 1):
        print(f"=== SORTEIO POTE {i} ===")

        # Embaralhar países do pote atual
        paises_embaralhados = list(pote)
        random.shuffle(paises_embaralhados)

        for pais in paises_embaralhados:
            confederacao_pais = encontrar_confederacao(pais)

            # Encontrar grupos disponíveis para este país
            grupos_disponiveis = []

            for nome_grupo in grupos.keys():
                grupo_atual = grupos[nome_grupo]
                contador_atual = contadores_grupos[nome_grupo]

                # Verificar se o grupo já tem 6 times (máximo)
                if len(grupo_atual) >= 6:
                    continue

                # Verificar restrição de confederação
                max_permitido = 2  # Padrão
                if confederacao_pais == 'OFC (Oceania)':
                    max_permitido = 1

                if contador_atual[confederacao_pais] >= max_permitido:
                    continue

                # Verificar restrições específicas por confederação
                if confederacao_pais == 'UEFA (Europa)':
                    if contador_atual[confederacao_pais] >= 2:
                        continue

                grupos_disponiveis.append(nome_grupo)

            # Sortear entre os grupos disponíveis
            if grupos_disponiveis:
                grupo_sorteado = random.choice(grupos_disponiveis)
                grupos[grupo_sorteado].append(pais)
                contadores_grupos[grupo_sorteado][confederacao_pais] += 1

                print(f"{pais} → {grupo_sorteado}")
            else:
                print(f"ERRO: Nenhum grupo disponível para {pais}")

        print()

    # 6. EXIBIR GRUPOS FINAIS
    print("=== GRUPOS FINAIS COPA 2026 ===")
    print("Formato: 8 grupos com 6 seleções cada\n")

    for grupo_nome, times in grupos.items():
        print(f"{grupo_nome}:")
        for i, time in enumerate(times, 1):
            confederacao = encontrar_confederacao(time)
            print(f"  {i}. {time} ({confederacao.split()[0]})")
        print(f"  Total: {len(times)} seleções\n")

    # 7. ESTATÍSTICAS DO CHAVEAMENTO
    print("=== ESTATÍSTICAS DO CHAVEAMENTO ===")

    # Contar seleções por confederação nos grupos
    estatisticas_confederacao = {}
    for confederacao in paises_por_confederacao.keys():
        estatisticas_confederacao[confederacao] = 0

    for grupo_times in grupos.values():
        for time in grupo_times:
            confed = encontrar_confederacao(time)
            if confed in estatisticas_confederacao:
                estatisticas_confederacao[confed] += 1

    for confederacao, quantidade in estatisticas_confederacao.items():
        print(f"{confederacao}: {quantidade} seleções")

    return grupos

# Executar o chaveamento
if __name__ == "__main__":
    grupos_finais = chaveamento_copa_2026()
```
